```python
# Importações:
import numpy as np
from os import path
import librosa as lb
from tqdm import tqdm
from pytube import YouTube
from pydub import AudioSegment
from spleeter.separator import Separator
from IPython.display import Audio, display
from google.colab import files
```

```python
def audioDownload(link):
  '''
      Função construida para se baixar
  os vídeos do youtube como audio.
  '''
  yt = YouTube(link)# Criando o objeto youtube com a url
  audio_stream = yt.streams.filter(only_audio=True).order_by('abr').desc().first()
  print('Download executado com sucesso')
  return audio_stream.download()

def converter_para_wav(audio_path, formato_saida='wav'):
  '''
    Função que converte o arquivo de audio para wav.
  O arquivo que se queria utilizar pode não estar num
  formato adequado, então essa função converte o arquivo
  de audio para um formato mais adequado.
  '''
  # Carregar o arquivo de áudio
  audio = AudioSegment.from_file(audio_path)

  # Converter para WAV (formato padrão suportado pelo SpeechRecognition)
  novo_caminho = f"{audio_path.rsplit('.', 1)[0]}.{formato_saida}"
  audio.export(novo_caminho, format=formato_saida)
  return novo_caminho

def separador(path, stems):
    '''
        Função construida com os comandos
    para a separação das tracks.
    '''
    modelo_separador = Separator(f'spleeter:{stems}')
    modelo_separador.separate_to_file(path,'')

def audio_extract(diretorio, play=True, sr=44100):
    '''
        Função que, dado o diretório da música,
    extrai a waveform da mesma e sua taxa de
    amostragem.
    '''

    y, sr = lb.load(diretorio, sr=sr)
    display(Audio(data=y,rate=sr))
    return y, sr
```

```python
# Entrada:
url = input('Insira a url do arquivo:\n')

# Baixando musica:
path = audioDownload(url)

# Convertendo de webm para wav:
path_wav = converter_para_wav(path)

# Splitando as tracks:
separador(path_wav, '5stems')
```
